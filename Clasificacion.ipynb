```python
# Lista para almacenar los DataFrames
for u in range (len(paths)):
    # Valores para la columna 'label'
    num_users = len(paths)
    labels = [0] * num_users
    user_to_classify = u # Índice del usuario a autenticar; CAMBIAR según corresponda
    print("USUARIO"+str(u+1) +"\n")
    labels[user_to_classify] = 1

    dataframes = []
    num_files = []
    
    # Cargar cada archivo CSV y añadirlo a la lista de DataFrames
    for y, path in enumerate(paths):
        df = pd.read_csv(path).T
        df['label'] = labels[y]
        dataframes.append(df)
        num_files.append(df.shape[0])

    # Determinar el número mínimo de columnas
    min_rows = min(num_files)

    # Truncar todas las tablas al número mínimo de columnas, incluyendo 'label'
    for y in range(len(dataframes)):
        dataframes[y] = dataframes[y].iloc[:min_rows, :].copy()  # Truncar las columnas de datos
        dataframes[y]['label'] = labels[y]  # Asegurarse de incluir la columna 'label'
     
    #Concatenar todos los DataFrames en uno solo
    dataset = pd.concat(dataframes, axis=0).reset_index(drop=True) 
    dataset = dataset.fillna(0)

    print(dataset.shape)
    
    X = dataset.drop('label',axis=1)
    y = dataset['label']

    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
    
    
   #Random Forest
    model_rf = RandomForestClassifier(n_estimators=200, max_depth=25, random_state=42)
    #Entrenar el modelo con el conjunto de entrenamiento
    model_rf.fit(X_train, y_train)
    #Predecir las etiquetas en el conjunto de prueba
    y_pred = model_rf.predict(X_test)
    
    #SVM
    #model_SVM = SVC(kernel='rbf', gamma='scale', random_state=42)
    #model_SVM.fit(X_train,y_train)
    #y_pred = model_SVM.predict(X_test)


    # k-NN
    #model_knn = KNeighborsClassifier(n_neighbors=20)
    #model_knn.fit(X_train, y_train)
    #y_pred = model_knn.predict(X_test)

    #LDA
    
    #model_lda = LinearDiscriminantAnalysis() # Inicializar el modelo LDA
    #model_lda.fit(X_train, y_train) #Entrenar el modelo
    #y_pred = model_lda.predict(X_test)
    
    # Evaluar la precisión del modelo
    accuracy = metrics.accuracy_score(y_test, y_pred)
    print(f"Accuracy del modelo RF =: {accuracy:.4f}")
    
    # Matriz de confusión
    cnf_matrix_RF = metrics.confusion_matrix(y_test, y_pred)
    TN = cnf_matrix_RF[0][0]
    TP = cnf_matrix_RF[1][1]
    FP = cnf_matrix_RF[0][1]
    FN = cnf_matrix_RF[1][0]

```
